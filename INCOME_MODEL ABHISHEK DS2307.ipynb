```python
data.skew()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.corr().abs(),annot=True)
```

```python
data.drop('Sex',axis=1, inplace=True)
```

```python
"""sex and relationship is multicolinear so droping less target related data sex from dataset."""
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
"""for cross verification using VIF for feature selection"""
```

```python
X=data.drop('Income', axis=1)
y=data['Income']
```

```python
std=StandardScaler()

scld=std.fit_transform(X)
```

```python
vif=pd.DataFrame()

vif['score']=[variance_inflation_factor(scld,i)for i in range(scld.shape[1])]
vif['Name']=X.columns
```

```python
vif
```

```python
"""Dataset is now No feature is multicolinear"""
```

# First model -KNN


```python
# We are having good no of data so choosing 70% for training and 30% for testing.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
kn=KNeighborsClassifier()
best_acc=0
best_random_state=0

for i in range(100,200):
    x_train,x_test,y_train,y_test=train_test_split(scld,y, test_size=0.30, random_state=i)
    

    kn.fit(x_train,y_train)

    pred=kn.predict(x_train)
    acc=accuracy_score(pred,y_train)

    if acc > best_acc:
        best_acc=acc
        best_random_state=i

print(best_acc)
print(best_random_state)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scld,y, test_size=0.30, random_state=174)
kn.fit(x_train,y_train)

ypred=kn.predict(x_train)
acc_train=accuracy_score(ypred,y_train)
print('Best training accuracy is',acc_train)

pred=kn.predict(x_test)
acc_test=accuracy_score(pred,y_test)
print('Best test accuracy is',acc_test)
print(confusion_matrix(y_test,pred))
```

```python
#At random state accuracy is 88% and test accuracy is 82 %.
#Confusion matrix is giving quite good answers.
#Yet score is good but trying ,if accuracy increases 
```

# Second model is DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
dt=DecisionTreeClassifier()
best_acc=0
best_random_state=0

for i in range(1,50):
    x_train,x_test,y_train,y_test=train_test_split(scld,y, test_size=0.30, random_state=i)
    

    dt.fit(x_train,y_train)

    pred=dt.predict(x_train)
    acc=accuracy_score(pred,y_train)

    if acc > best_acc:
        best_acc=acc
        best_random_state=i

print(best_acc)
print(best_random_state)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scld,y, test_size=0.30, random_state=1)
ypred=dt.predict(x_train)
acc_train=accuracy_score(ypred,y_train)
print('Best training accuracy is',acc_train)

pred=dt.predict(x_test)
acc_test=accuracy_score(pred,y_test)
print('Best test accuracy is',acc_test)
print(confusion_matrix(y_test,pred))
```

```python
#Decision Tree giving very good results and confusion matrix prediction is also very good
```

# Third is logisticRegression

```python
from sklearn.linear_model import LogisticRegression

ln=LogisticRegression()
best_acc=0
best_random_state=0

for i in range(1,800):
    x_train,x_test,y_train,y_test=train_test_split(scld,y, test_size=0.25, random_state=i)
    

    ln.fit(x_train,y_train)

    pred=ln.predict(x_train)
    acc=accuracy_score(pred,y_train)

    if acc > best_acc:
        best_acc=acc
        best_random_state=i

print(best_acc)
print(best_random_state)

```

```python
x_train,x_test,y_train,y_test=train_test_split(scld,y, test_size=0.25, random_state=218)
ln.fit(x_train,y_train)

ypred=ln.predict(x_train)
acc_train=accuracy_score(ypred,y_train)
print('Best training accuracy is',acc_train)

pred=ln.predict(x_test)
acc_test=accuracy_score(pred,y_test)
print('Best test accuracy is',acc_test)
print(confusion_matrix(y_test,pred))
```
