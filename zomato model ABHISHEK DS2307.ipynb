```python
from scipy.stats import zscore
data=new[['Average Cost for two']]

z=np.abs(zscore(data))
z
```

```python
np.where(z>3)
```

```python
data=new[(z<3).all(axis=1)] #avg cost for 2 is highly right skewed so treating it
data.shape
```

```python
data['Average Cost for two']=np.sqrt(data['Average Cost for two'])
data['Votes']=np.sqrt(data['Votes'])
```

```python
data.skew()
```

```python
data['Average Cost for two']=np.cbrt(data['Average Cost for two'])
data['Votes']=np.cbrt(data['Votes'])
```

```python
data.skew()
```

```python
plt.figure(figsize=(15,20))
p=1

for column in data:
    if p<=22:
        ax=plt.subplot(5,5,p)
        sns.distplot(data[column])
        
    p+=1
plt.show()
```

```python
data.drop_duplicates()  #droping duplicates
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('Average Cost for two', axis=1)
y=data['Average Cost for two']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
ln=LinearRegression()
best_acc=0
best_random_state=0

for i in range(1,1200):
    x_train,x_test,y_train,y_test=train_test_split(X,y, test_size=0.25, random_state=i)
    

    ln.fit(x_train,y_train)

    pred=ln.predict(x_train)
    acc=r2_score(pred,y_train)

    if acc > best_acc:
        best_acc=acc
        best_random_state=i

print(best_acc)
print(best_random_state)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y, test_size=0.25, random_state=407)
ln.fit(x_train,y_train)

ypred=ln.predict(x_train)
acc_train=r2_score(ypred,y_train)
print('Best training accuracy is',acc_train)

pred=ln.predict(x_test)
acc_test=r2_score(pred,y_test)
print('Best test accuracy is',acc_test)
```

```python
A=data.drop('Price range', axis=1)
b=data['Price range']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
kn=KNeighborsClassifier()
best_acc=0
best_random_state=0

for i in range(1,1500):
    x_train,x_test,y_train,y_test=train_test_split(A,b, test_size=0.30, random_state=i)
    

    kn.fit(x_train,y_train)

    pred=kn.predict(x_train)
    acc=accuracy_score(pred,y_train)

    if acc > best_acc:
        best_acc=acc
        best_random_state=i

print(best_acc)
print(best_random_state)
```

```python
x_train,x_test,y_train,y_test=train_test_split(A,b, test_size=0.30, random_state=424)
kn.fit(x_train,y_train)

ypred=kn.predict(x_train)
acc_train=accuracy_score(ypred,y_train)
print('Best training accuracy is',acc_train)

pred=kn.predict(x_test)
acc_test=accuracy_score(pred,y_test)
print('Best test accuracy is',acc_test)
print(confusion_matrix(y_test,pred))
```
