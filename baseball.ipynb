```python
HSV=q3['SV']+1.5*iqr['SV']
print(HSV)

find= np.where(df['SV']>HSV)
df=df.drop(df.index[find])
df.shape

df.reset_index()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#as dataset is having differnt scales.For good understanding making all on a same scale.
```

```python
std=StandardScaler()
```

```python
X= df.drop('W', axis=1)
y=df['W']
```

```python
sclaed=std.fit_transform(X)
```

```python
#cross verifying for feature selection using VIF
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
VIF=pd.DataFrame()
```

```python
VIF['vif']=[variance_inflation_factor(sclaed,i) for i in range(sclaed.shape[1])]
VIF['NAME']=X.columns
```

```python
VIF
```

```python
X=X.drop('3B',axis=1)
```

```python
#VIF score is high so droping 3B
```

```python
sclaed=std.fit_transform(X)
```

```python
VIF=pd.DataFrame()
```

```python
VIF['vif']=[variance_inflation_factor(sclaed,i) for i in range(sclaed.shape[1])]
VIF['NAME']=X.columns
```

```python
VIF
```

```python
#vif is under range. ALL feature are important
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y, test_size=0.25,random_state=450)
```

```python
#1st model is linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
#train score is 88% and test score is 78%
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
ad=AdaBoostRegressor()
```

```python
ad.fit(x_train,y_train)
```

```python
ad.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
best_param={'n_estimators':range(1,30),
           'learning_rate':np.arange(0.1,0.5),
           }
```

```python
GD=GridSearchCV(ad,param_grid=best_param)
```

```python
GD.fit(x_train,y_train)
```

```python
GD.best_params_
```

```python
add=AdaBoostRegressor(learning_rate= 0.1, n_estimators= 3)
```

```python
add.fit(x_train,y_train)
```

```python
add.score(x_train,y_train)
```

```python
add.score(x_test,y_test)
```
